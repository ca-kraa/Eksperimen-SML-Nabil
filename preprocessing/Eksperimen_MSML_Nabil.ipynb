## Deskripsi Dataset

Dataset ini digunakan untuk melakukan klasifikasi penyakit jantung berdasarkan beberapa atribut kesehatan pasien. Data terdiri dari berbagai fitur medis yang berkaitan dengan kondisi kesehatan seseorang, seperti:

* Usia (`age`)
* Jenis kelamin (`sex`)
* Tekanan darah (`trestbps`)
* Kadar kolesterol (`chol`)
* Detak jantung maksimum (`thalach`)
* Kadar gula darah (`fbs`)
* Jenis nyeri dada (`cp`)
* Dan beberapa atribut medis lainnya.

Dataset memiliki target klasifikasi berupa:

* `0` → Tidak memiliki penyakit jantung
* `1` → Memiliki penyakit jantung
